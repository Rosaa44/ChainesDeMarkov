2.Soit X la variable aléatoire qui indique la longueur de la sequence I.On a donc
P(X=k)=(q^(k-1))*p avec q la probabilité de rester infecté si on est infecté au temps précédent donc q=P(In|In-1)=0.93 et p=1-q=0.07. Donc X suit une loi géométrique de paramètre p alors l'esperance de X est E[X]=1/p=1/0.07=14.286.
Donc la longueur d'une sequence I est en moyenne 14.286 ce qui semble bien être notre résultat dans la simulation.
